```python
def compute_intersect(id_list, headername):

    def search(word):
        val = list()
        for x in training_material:
            if word in x:
                yield x   

    def flatten(x):
        for y in x:
            for z in y:
                yield z

    x=[search(word) for word in id_list]

    x2 = list()
    for value in x:
        x2.append(flatten(value))


    def intersection(l1, l2): 
        return [value for value in l1 if value in l2]

    x3 = list()
    for i in range(0,len(x2)):
        x3.append([value for value in x2[i]])
    intersect = x3[0]
    for i in range(1,len(x3)):
        intersect = intersection(x3[i], intersect)
    print("\n------------")
    print('Intersect for ' + headername+':')
    print(intersect)
    return intersect
```
