```python
import pandas as pd
import numpy as np
pd.options.display.max_colwidth = 500
```

```python
candidates = pd.read_csv('C:/Users/D072202/Desktop/testmappings.csv',sep="\t",header=None)
test_final = pd.read_csv('C:/Users/D072202/Desktop/result_data2/broadwalkembedder_2019_05_28_18_36_11_566479/possible_matches.csv-strcombined.csv', index_col=['Unnamed: 0'])
test_final_ids = pd.read_csv('C:/Users/D072202/Desktop/result_data2/broadwalkembedder_2019_05_28_18_36_11_566479/possible_matches.csv-strcombined_ids.csv', index_col=['Unnamed: 0'])
```

```python
c = None
for index, row in candidates.iterrows():
    tmp = test_final_ids.loc[(test_final_ids.src_id==row[0]) & (test_final_ids.tgt_id==row[1])][['src_id','tgt_id']].head(1)
    if c is None:
        c = tmp
    else:
        c = c.append(tmp, ignore_index=False)
c
```

```python
c=c.merge(test_final, right_index=True, left_index=True)
#c = c.loc[c.label==1]
c = c[[col for col in c.columns if col not in ['src_id','tgt_id']]]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from pandas_ml import ConfusionMatrix
```

```python
train_final = pd.read_csv('C:/Users/D072202/Desktop/result_data2/simpletriples infer_2019_05_28_12_47_52_414900/final_trainset.csv-strcombined.csv', index_col=['Unnamed: 0'])
len(train_final.loc[train_final.label==1])
```

```python
train_final = train_final.loc[(train_final.syntactic_diff < 0.4) & (train_final.plus_diff < 0.9)]
len(train_final.loc[(train_final.syntactic_diff < 0.4) &(train_final.label==0) & (train_final.plus_diff < 0.9)])
train_final.loc[train_final.label==1].head()
```

```python
x_train = train_final.loc[:, [col for col in train_final.columns if not col in ['label', 'syntactic_diff','plus_diff']]]
y_train = train_final['label']

x_test1 = c.loc[:, [col for col in c.columns if not col in ['label','syntactic_diff','plus_diff']]]
y_test1 = c['label']
```

```python
model = RandomForestClassifier(n_estimators=10, max_depth=5, random_state=0) #LogisticRegression(solver='lbfgs')
model = model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test1)
result = classification_report(np.array(y_test1), prediction, target_names=['false','true'])
print("Results on simple test:")
print(result)
print(ConfusionMatrix( np.array(y_test1), prediction))
print("\n\n--------------------------------------------------------------\n")
```
