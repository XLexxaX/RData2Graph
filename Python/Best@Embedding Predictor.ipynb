```python
import pandas as pd
import numpy as np
```

```python
test_final = pd.read_csv('C:/Users/D072202/Desktop/result_data2/broadwalkembedder_2019_05_28_18_36_11_566479/possible_matches.csv-strcombined.csv', index_col=['Unnamed: 0'])
test_final_ids = pd.read_csv('C:/Users/D072202/Desktop/result_data2/broadwalkembedder_2019_05_28_18_36_11_566479/possible_matches.csv-strcombined_ids.csv', index_col=['Unnamed: 0'])
```

```python
test_final['dist_to_mean'] = (test_final.src_tgt_angle.abs() - 0.75).abs()
```

```python
test_final['src_tgt_angle'] = test_final['src_tgt_angle'].abs()
test_final['src_tgt_angle'] = test_final['src_tgt_angle'].round(2) 
```

```python
test_final = test_final.loc[(test_final.src_tgt_angle>=0.59) & (test_final.src_tgt_angle<=0.79) & (test_final.src_tgt_veclen>5.0)]
test_final_ids = test_final_ids.iloc[test_final.index]
```

```python
src_ids = set(test_final_ids['src_id'])
tgt_ids = set(test_final_ids['tgt_id'])
```

```python
relevant_attributes = ['src_tgt_angle','plus_diff']
minmax = ['max','min']
test_final['score'] = 0.0

for src_id in src_ids:
    x = test_final.loc[test_final_ids.loc[test_final_ids.src_id==src_id].index]
    
    for i in range(len(relevant_attributes)):
        att = relevant_attributes[i]
        if minmax[i] == 'max':
            x.loc[x[att].isin(set(x.nlargest(2, att)[att].to_list())), 'score'] = x.loc[x[att].isin(set(x.nlargest(2, att)[att].to_list())), 'score'] + 1
        else:
            x.loc[x[att].isin(set(x.nsmallest(2, att)[att].to_list())), 'score'] = x.loc[x[att].isin(set(x.nsmallest(2, att)[att].to_list())), 'score'] + 1
    test_final.loc[test_final_ids.loc[test_final_ids.src_id==src_id].index, [c=='score' for c in test_final.columns]] = pd.DataFrame(x['score'])
test_final.head()
```

```python
for tgt_id in tgt_ids:
    x = test_final.iloc[test_final_ids.loc[test_final_ids.tgt_id==tgt_id].index]
    
    for i in range(len(relevant_attributes)):
        att = relevant_attributes[i]
        if minmax[i] == 'max':
            x.loc[x[att].isin(set(x.nlargest(2, att)[att].to_list())), 'score'] = x.loc[x[att].isin(set(x.nlargest(2, att)[att].to_list())), 'score'] + 1
        else:
            x.loc[x[att].isin(set(x.nsmallest(2, att)[att].to_list())), 'score'] = x.loc[x[att].isin(set(x.nsmallest(2, att)[att].to_list())), 'score'] + 1
    test_final.iloc[test_final_ids.loc[test_final_ids.tgt_id==tgt_id].index, [c=='score' for c in test_final.columns]] = pd.DataFrame(x['score'])
test_final.head()
```

```python

with open('C:/Users/D072202/Desktop/testmappings.csv', mode="w+") as f:
    for src_id in src_ids:
        x = test_final.loc[test_final_ids.loc[test_final_ids.src_id==src_id].index]
        #x = x.loc[x.plus_diff < 0.8]
        if x['score'].max() > 1.0:
            y = test_final_ids.loc[x.loc[x.score==x['score'].max()].index]
            for index, row in y.iterrows():
                    f.write(row.src_id + " \t " + row.tgt_id)
                    f.write("\t" + str(test_final.loc[index].label)+"\n")
```
