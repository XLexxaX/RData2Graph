```python
import uuid

manufacturers = dict()

def remove_quotes(line):
    newline = []
    for term in line:
        if "\"" in term:
            term = term[1:-1]
        newline = newline + [term]
    return newline
```

```python
ctr = 0
with open("C:/Users/D072202/Downloads/amazon_google/Amazon.csv", "r", encoding="UTF-8") as lines:
    for line in lines:
        if ctr > 0:
            line = line.split(",")
            line = remove_quotes(line)
            manufacturers[line[3]] =str(uuid.uuid4())
        ctr = ctr + 1
```

```python
manufacturerfile = open("C:/Users/D072202/Downloads/amazon_google/Amazon_manufacturer.csv","w")

manufacturerfile.write("Amazon_manufacturer.id,Amazon_manufacturer.title\n")
for manufacturer in manufacturers.keys():
    line = remove_quotes(line)
    manufacturerfile.write(manufacturers[manufacturer] + "," + manufacturer +"\n")

manufacturerfile.close()
```

```python
productsfile = open("C:/Users/D072202/Downloads/amazon_google/Amazon_products.csv","w")

ctr = 0
with open("C:/Users/D072202/Downloads/amazon_google/Amazon.csv", "r", encoding="UTF-8") as lines:
    for line in lines:
        if ctr > 0:
            line = line.split(",")
            line = remove_quotes(line)
            productsfile.write(line[0] + "," + line[1] + "," + line[2] + "," + manufacturers[line[3]] + "," + line[4])
        else:
            line = line.split(",")
            line = remove_quotes(line)
            productsfile.write("Amazon_products."+line[0] + "," + "Amazon_products."+line[1] + "," + "Amazon_products."+line[2] + "," + "Amazon_manufacturer."+line[3] + "," + "Amazon_products."+line[4])
            ctr = ctr + 1
productsfile.close()
```
