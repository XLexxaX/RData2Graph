```python
matchings.loc[matchings.total_score_y.isna(),'total_score_y'] = matchings.loc[matchings.total_score_y.isna(),'total_score_x']
matchings.loc[matchings.total_score_y.isna(),'syntactic_diff_y'] = matchings.loc[matchings.total_score_y.isna(),'syntactic_diff_x']
matchings.loc[matchings.total_score_y.isna(),'src_tgt_angle_y'] = matchings.loc[matchings.total_score_y.isna(),'src_tgt_angle_x']
matchings.head()
```
